```python
df = pd.read_csv('/content/Problem3.csv')
df.head()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()
for col in categorical_cols:
    n_categories = df[col].nunique()
    print(f'Number of categories in {col}: {n_categories}')


ordinal_encoder = OrdinalEncoder()
encoded_categorial_cols = ordinal_encoder.fit_transform(df[categorical_cols])
```

```python
import numpy as np

# Data
Y = np.array([50, 70, 80, 85])

# Initialize f0 as the average of Y
f0 = np.mean(Y)

# Calculate residuals
residuals = Y - f0

# Calculate sum of residuals
sum_of_residuals = np.sum(residuals)

# Number of residuals
number_of_residuals = len(Y)

# Lambda
lambda_value = 0

# Calculate similarity score
similarity_score = (sum_of_residuals ** 2) / (number_of_residuals + lambda_value)

# Print the results
print("Initialized f0 value:", f0)
print("Sum of Residuals:", sum_of_residuals)
print("Number of Residuals:", number_of_residuals)
print("Similarity Score at Root Node:", similarity_score)
```

```python
def similarity_score(y_true, y_pred, lamda):
    residuals = y_true - y_pred
    similarity_score = np.sum(residuals)**2 / (lamda + len(residuals))
    return similarity_score
```

```python
def split_and_score(df, condition, lamda, sc_root):
    df_left = df[df['X'] < condition]
    df_right = df[df['X'] >= condition]
    y_pred_right = np.full(len(df_right), np.mean(df_right['Y']))
    y_pred_left = np.full(len(df_left), np.mean(df_left['Y']))
    sc_left = similarity_score(df_left['Y'].values, y_pred_left, lamda)
    sc_right = similarity_score(df_right['Y'].values, y_pred_right, lamda)


    gain = sc_root - (len(df_left) / len(df)) * sc_left - (len(df_right) / len(df)) * sc_right

    return sc_left, sc_right, gain
```
